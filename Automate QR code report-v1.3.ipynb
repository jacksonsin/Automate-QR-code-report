# Generate New Excel file with Today's date

```python
from datetime import datetime, timedelta
start_date = []
end_date = []
new_date = []
new_date = pd.to_datetime(result[0]).dt.date

for x in result[0]:
    dt = datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S')
    dt1 = dt - timedelta(days=dt.weekday())
    start_date.append(dt1.strftime("%d/%m/%Y"))
    dt2 = dt1 + timedelta(days=6)
    end_date.append(dt2.strftime("%d/%m/%Y"))
```
